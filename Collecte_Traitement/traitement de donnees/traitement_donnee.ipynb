```python
import numpy as np
import pandas as pd
```

```python
data_file = pd.read_csv('personnes.csv')
```

```python
df = pd.DataFrame(data=data_file)
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.loc[df['email'].duplicated(keep=False),:]
```

```python
VALID_COUNTRIES = ['France', 'Côte d\'ivoire', 'Madagascar', 'Bénin', 'Allemagne', 'USA']
mask = ~df['pays'].isin(VALID_COUNTRIES)
df.loc[mask, 'pays'] = np.NaN
```

```python
df
```

```python
df['email'] = df['email'].str.split(',', n=1, expand=True)[0]
```

```python
df
```

```python
df['taille'] = df['taille'].str[:-1]
df['taille'] = pd.to_numeric(df['taille'], errors='coerce')
```

```python
df
```

```python
df.loc[df['taille'].isnull(), 'taille'] = df['taille'].mean()
```

```python
df
```

```python
df['date_naissance'] = pd.to_datetime(df['date_naissance'], format='%d/%m/%Y', errors='coerce')
```
